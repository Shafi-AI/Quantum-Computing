```python
%matplotlib inline
# Importing standard Qiskit libraries and configuring account
from qiskit import QuantumCircuit, execute, Aer, IBMQ
from qiskit.compiler import transpile, assemble
from qiskit.tools.jupyter import *
from qiskit.visualization import *
# Loading your IBM Q account(s)
provider = IBMQ.load_account()
```

```python
from qiskit.visualization import plot_histogram , plot_bloch_vector
from math import sqrt, pi 
```

```python
%config InlineBackend.figure_format = 'svg' # Makes the images look nice
```

```python
# QuantumCircuit is used to Store Circuits in Quantum Computing 
qc = QuantumCircuit(1)
```

```python
In Quantum Circuits, qubits always start out in the state |0> - Zero State. We can use initialize() method to transform in to 
any state. 
```

```python
initial_state = [0,1]
qc.initialize(initial_state,0)
qc.draw()
```

We can use any simulator to view the resulting state of our Qubit. Here using statevector simulator to view. 


```python
backend = Aer.get_backend("statevector_simulator")
```

To run our circuit use execute() method by passing the circuit and backend objects, and use result() method to get the result.


```python
result = execute(qc, backend).result()
```

```python
# We can use get_statevector() method to get the state vector output . 
out_state = result.get_statevector()
print(out_state)
```

```python
qc1 =  QuantumCircuit(1)
zero_state  = execute(qc1,backend).result().get_statevector()
print(zero_state)
# Here we get |0> vector because we did not initialize, by default it is zero vector 
```

```python
Measure our Qubit and see the result in real Quantum Computer. 
```

```python
qc.measure_all()
qc.draw()
```

```python
result = execute (qc, backend).result()
counts = result.get_counts()
plot_histogram(counts)
```

```python
resultFor0Qubit =  execute(qc1, backend).resultult()
ZeroCounts = resultFor0Qubit.get_counts()
plot_histogram(ZeroCounts)
```

StateVector: In Quantum Physics we use Statevectors to describe the state of the system. In otherwords, collection of numbers in a vector called a Statevector. Each element in the statevector is probability of occuring the event. 

Explording Qubits with Qiskit

```python
from qiskit import QuantumCircuit , execute, Aer
from qiskit.visualization import plot_histogram , plot_bloch_vector
from math import sqrt, pi 
```

```python
%config InlineBackend.figure_format = 'svg'
```

```python
qc = QuantumCircuit(1)
```

```python
initial_state = [0,1]
qc.initialize(initial_state, 0)
qc.draw()
```

```python
backend =  Aer.get_backend('statevector_simulator')
```

```python
result =  execute(qc,backend).result()
```

```python
output_state =  result.get_statevector()
print(output_state)
```

```python
qc.measure_all()
qc.draw()
```

```python
result = execute(qc, backend).result()
counts = result.get_counts()
plot_histogram(counts)
```

```python
qc = QuantumCircuit(1)
initial_state =  [0.+1.j/sqrt(2), 1/sqrt(2) + 0.j]
qc.initialize(initial_state, [0])
qc.draw()
```
